```python
#If there is any numeric it wont be considered
print("MEAN:\n\n",df.mean(numeric_only=True))
print("\nMEDIAN:\n\n",df.median(numeric_only=True))
print("\nMODE:\n\n",df.mode().iloc[0])
print("\nSTANDARD DEVIATION:\n\n",df.std(numeric_only=True))
print("\nVARIANCE:\n\n",df.var(numeric_only=True))


print("\nRANGE:\n\n",df.max(numeric_only=True)-df.min(numeric_only=True))
print("\nSKEWNESS:\n\n",df.skew(numeric_only=True))
print("\nKURTOSIS:\n\n",df.kurt(numeric_only=True))
```

```python
from scipy import stats

# Access the 'Life expectancy '
X = df['infant deaths']
# population mean
population_mean = 1.50
# perform one sample test
t_stat, p_value = stats.ttest_1samp(X, population_mean)

print(f"T-static:{t_stat}")
print(f"p-vales:{p_value}")
```

```python
import numpy as np
from scipy import stats
```

```python
#Sample mean and standard error for infant deaths
sample_mean=np.mean(df['infant deaths']) # Pass the DataFrame column to np.mean()
standard_error=stats.sem(df['infant deaths']) # Pass the DataFrame column to stats.sem()
# Compute 95% confidence interval for infant deaths
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)
print(f"95% Confidence Interval for infant deaths: {confidence_interval}")
```

```python
import statsmodels.api as sm
```

```python
# Define independent variable (add constant for intercept)
X = sm.add_constant(df['infant deaths'])
#define dependent variable
#Check for typos and ensure that the column 'Life expectancy' exists.
y = df['Life expectancy ']
#fit linear regression model
model = sm.OLS(y, X).fit()
#print model summary
print(model.summary())
```
